```python
import random
import time

# ---------------------------
# Ambiente
# ---------------------------
class Ambiente:
    def __init__(self, n_salas, n_sujas):
        self.n_salas = n_salas
        self.salas = [0] * n_salas  # 0 = limpa, 1 = suja

        # sorteia salas sujas
        sujas = random.sample(range(n_salas), n_sujas)
        for i in sujas:
            self.salas[i] = 1

        # posição inicial do aspirador
        self.pos = random.randint(0, n_salas - 1)

    def esta_suja(self, pos):
        return self.salas[pos] == 1

    def limpar(self):
        self.salas[self.pos] = 0

    def mover_esquerda(self):
        if self.pos > 0:
            self.pos -= 1

    def mover_direita(self):
        if self.pos < self.n_salas - 1:
            self.pos += 1


# ---------------------------
# Interface
# ---------------------------
def mostrar_ambiente(amb, modo="base"):
    print("\nSALAS:", list(range(amb.n_salas)))

    asp = [" "] * amb.n_salas
    asp[amb.pos] = "A"
    print("Aspirador:", asp)

    sujeira = []
    for i in range(amb.n_salas):
        if modo == "onisciente":
            sujeira.append("S" if amb.salas[i] == 1 else ".")
        else:
            if i == amb.pos:
                sujeira.append("S" if amb.salas[i] == 1 else ".")
            else:
                sujeira.append("?")
    print("Sujeira:  ", sujeira)


# ---------------------------
# Controlador Manual
# ---------------------------
def controlador_manual(amb):
    while True:
        mostrar_ambiente(amb, modo="onisciente")
        if sum(amb.salas) == 0:
            print("Tudo limpo (manual).")
            break
        acao = input("Ação (l=limpar, e=esq, d=dir): ")

        if acao == "l":
            amb.limpar()
        elif acao == "e":
            amb.mover_esquerda()
        elif acao == "d":
            amb.mover_direita()


# ---------------------------
# Controlador Base
# ---------------------------
def controlador_base(amb):
    direcao = 1  # 1 = direita, -1 = esquerda
    sem_sujeira = 0

    while True:
        mostrar_ambiente(amb, modo="base")
        time.sleep(0.5)

        # limpa se estiver sujo
        if amb.esta_suja(amb.pos):
            amb.limpar()
            sem_sujeira = 0
        else:
            sem_sujeira += 1

        # critério de parada (heurístico correto)
        if sem_sujeira >= amb.n_salas * 2:
            print("Acredita que terminou (base).")
            break

        # muda direção nas bordas
        if amb.pos == amb.n_salas - 1:
            direcao = -1
        elif amb.pos == 0:
            direcao = 1

        # movimento
        if direcao == 1:
            amb.mover_direita()
        else:
            amb.mover_esquerda()


# ---------------------------
# Controlador Onisciente
# ---------------------------
def controlador_onisciente(amb):
    while True:
        mostrar_ambiente(amb, modo="onisciente")
        time.sleep(0.5)

        # limpa se necessário
        if amb.esta_suja(amb.pos):
            amb.limpar()
            continue

        # verifica se acabou
        if sum(amb.salas) == 0:
            print("Tudo limpo (onisciente).")
            break

        # escolhe sujeira mais próxima (guloso)
        sujeiras = [i for i in range(amb.n_salas) if amb.salas[i] == 1]
        alvo = min(sujeiras, key=lambda x: abs(x - amb.pos))

        if amb.pos < alvo:
            amb.mover_direita()
        elif amb.pos > alvo:
            amb.mover_esquerda()


# ---------------------------
# Main
# ---------------------------
def main():
    n = int(input("Número de salas (max 10): "))
    n = min(n, 10)

    sujas = int(input("Número de salas sujas: "))
    sujas = min(sujas, n)

    amb = Ambiente(n, sujas)

    modo = input("Modo ('m' manual/ 'b' base/ 'o' onisciente): ")

    if modo == "m":
        controlador_manual(amb)
    elif modo == "b":
        controlador_base(amb)
    elif modo == "o":
        controlador_onisciente(amb)
    else:
        print("Modo inválido")


if __name__ == "__main__":
    main()
```
